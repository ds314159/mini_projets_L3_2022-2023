```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

# Plotting Epidemic Models

## Plotting basic SIR Model

```python
dataBS = pd.read_csv('basicSirEvolution.csv')

sline = plt.plot("t","S","",data=dataBS,color="blue",linewidth=2, label="Susceptibles")
iline = plt.plot("t","I","",data=dataBS,color="red",linewidth=2, label="Infected")
rline = plt.plot("t","R","",data=dataBS,color="green",linewidth=2, label="Recovered")
plt.title("Dynamics of transmission over T, Basic SIR")
plt.xlabel("Time Steps")
plt.ylabel("Population")
plt.legend()
```
